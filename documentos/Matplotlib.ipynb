## Representacion grafica de Datos.
Si la funcion de trazado se le da una matriz de datos, la usara como coordenadas en eje vertical, y utilizara el indice de cada punto de datos en el array como coordenada horizontal.

```python
plt.plot([1, 2, 5, 7, 8, 3, 1])
plt.show()
```

Tambien se puede proporcional dos matrices, una para el eje horizontal, y otra para el eje vertical

```python
plt.plot([-3, -1, 0, 4, 7], [1, 4, 6, 7, 8])
plt.show()
```

Pueden modificarse las longitudes de los ejes para que la figura no se vea tan ajustada

```python
plt.plot([-3, -1, 0, 4, 7], [1, 4, 6, 7, 8])
plt.axis([-4, 8, 0, 10]) # [Xmin, Xmax, Ymin, Ymax]
plt.show()
```

Se sigue el mismo procedimiento para pintar una funcion matematica

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2

plt.plot(x, y)
plt.show()
```

Tambien puede modificarse el estilo de la grafica para que contenga mas informacion.

```python
plt.plot(x, y)
plt.title("Square Funtion")
plt.xlabel("CAIM")
plt.ylabel("y =**2")
plt.grid(True)
plt.show()
```

Pueden superponerse graficas y cambiar el estilo de las funciones

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.plot(x, y)
plt.title("Square Funtion")
plt.xlabel("CAIM")
plt.ylabel("y =**2")
plt.grid(True)


plt.plot(x, y, 'b--', x, y2, 'r')
plt.show()
```

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.plot(x, y)
plt.title("Square Funtion")
plt.xlabel("CAIM")
plt.ylabel("y =**2")
plt.grid(True)


plt.plot(x, y, 'b--')
plt.plot(x, y2, 'g')
plt.show()
```

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.plot(x, y)
plt.title("Square Funtion")
plt.xlabel("CAIM")
plt.ylabel("y =**2")
plt.grid(True)


plt.plot(x, y, 'b--', label = "x**2")
plt.plot(x, y2, 'g', label = " x + 1")
plt.legend(loc = "best") # la situa en la mejor localizacion
plt.show()
```

Tambien pueden crearse dos graficas que no se superpongan. Estas graficas se organizan en un grid y se denominan subplots

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.subplot(1, 2, 1) #1 Rows, 2 columns, list subplot
plt.plot(x, y, 'r--')

plt.subplot (1, 2, 2) # 1 rows, 2 columns, 2nd Subplots
plt.plot(x, y2, 'g')

plt.show()
```

Para que las graficas no queden tan ajustadas, se puede hacer la figura mas grande.

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1) #1 Rows, 2 columns, list subplot
plt.plot(x, y, 'r--')
plt.title("Square Funtion", fontsize = 14)
plt.xlabel("CAIM", fontsize = 14)
plt.ylabel("y =**2", fontsize = 14)
plt.grid(True)


plt.subplot (1, 2, 2) # 1 rows, 2 columns, 2nd Subplots
plt.plot(x, y2, 'g')
plt.title("Lineal Funtion", fontsize = 14)
plt.xlabel("CAIM 2", fontsize = 14)
plt.ylabel("y =**2", fontsize = 14)
plt.grid(True)


plt.show()
```

## Scatter Plots

```python
from numpy.random import rand
x, y = rand(2, 100)
plt.scatter(x, y)
plt.show()
```

```python
from numpy.random import rand 
x, y = rand(2,100)
x2, y2 = rand(2, 100)
plt.scatter(x, y, c = 'red')
plt.scatter(x2, y2, c = 'purple')
plt.show()
```

## Histogramas 

```python
data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3,3,3]
plt.subplot(211)
plt.hist(data, bins = 10, rwidth = 0.8)

plt.subplot(212)
plt.hist(data, bins =[1, 1.5, 2, 2.5, 3], rwidth = 0.95)


plt.xlabel('Value')
plt.ylabel('Frecuency')

plt.show()
```
